```python
train_df=train_df.apply(pd.to_numeric, errors='coerce')
```

```python
for column in train_df.columns:
    train_df[column]=pd.to_numeric(train_df[column],downcast="float")
```

```python
for column in train_df.columns:
    counter=0
    first=True
    name=''
    for item in train_df[column]:
        if not( isinstance(item, float)):
            if first:
                name=column
                counter+=1
                first=False
            else :
                 counter+=1
    if name :
        print("{} : {} \n".format(name,counter))
```

```python
train_df.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
imp_mean = SimpleImputer(missing_values=np.nan, strategy='mean')
train_df=pd.DataFrame(imp_mean.fit_transform(train_df))
```

```python
train_df.isnull().sum()
```

```python
X_train=np.array(train_df.iloc[:,:-1])
y_train=np.array(train_df.iloc[:,-1:])
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
model = tf.keras.Sequential([
                tf.keras.layers.Dense(16,activation = tf.nn.relu),
                tf.keras.layers.Dense(12,activation = tf.nn.relu),                          
                tf.keras.layers.Dense(8,activation = tf.nn.relu),
                tf.keras.layers.Dense(1,activation = tf.nn.sigmoid)]
)
```

```python
from tensorflow.keras.optimizers import RMSprop
model.compile(optimizer=RMSprop(learning_rate=0.001),
              loss='binary_crossentropy',
              metrics = ['accuracy'])
```

```python
model.fit(X_train, y_train,epochs =1000,verbose=2)
```

```python
X_test=np.array(test_df)
```

```python
X_test.shape 
```

```python
predictions=model.predict(X_test)
```

```python
predictions=list(predictions)
```

```python
predictions
```

```python
for i in range(len(predictions)):
    if predictions[i]>0.7:
        predictions[i]=1
    else :
        predictions[i]=0
```

```python
predictions
```

```python
ids=test_df.index
```

```python
output=pd.DataFrame({'p_id':ids,'diabetes':predictions})
output.to_csv('submission.csv', index=False)
```
